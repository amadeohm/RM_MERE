```python
import pandas as pd
import numpy as np
```

```python
# Read CSV file
df = pd.read_csv("diabetes_dataset.csv")

# Determine 10% of the total number of rows to introduce missing values
num_nan = int(df.shape[0] * 0.1)

# Columns where missing values will be introduced
col_idx = ["employment_status", "gender", "family_history_diabetes", "bmi"]

# Types of missing values to insert in each corresponding column
# - np.nan: true missing value
# - "Unknown": categorical value representing "unknown" instead of NaN
# - 0: unrealistic BMI value (impossible in reality, default value)
inac = [np.nan, "Unknown", np.nan, 0]

# Loop through each target column and insert missing values randomly
df_nan = df.copy()
for col, val in zip(col_idx, inac):
    fila_idx = np.random.choice(df.index, size=num_nan, replace=False)
    df_nan.loc[fila_idx, col] = val

df_nan.to_csv('diabetes_dataset_nan.csv')
```
